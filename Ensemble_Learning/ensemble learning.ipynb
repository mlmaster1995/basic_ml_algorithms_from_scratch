```python
import numpy as np
import random
from random import random 
from collections import Counter

def weighted_choice(objects, weights):
    """ returns randomly an element from the sequence of 'objects', 
        the likelihood of the objects is weighted according 
        to the sequence of 'weights', i.e. percentages."""

    weights = np.array(weights, dtype=np.float64)
    sum_of_weights = weights.sum()
    # standardization:
    np.multiply(weights, 1 / sum_of_weights, weights)
    weights = weights.cumsum()
    x = random()
    for i in range(len(weights)):
        if x < weights[i]:
            return objects[i]
```

```python
def weighted_test(weights):
    """ returns randomly an element from the sequence of 'objects', 
        the likelihood of the objects is weighted according 
        to the sequence of 'weights', i.e. percentages."""

    weights = np.array(weights, dtype=np.float64)
    sum_of_weights = weights.sum()
    # standardization:
    np.multiply(weights, 1 / sum_of_weights, weights)
    weights = weights.cumsum()
    return weights

weights = [1/200, 1/600, 1/600, 1/600, 1/600, 1/1200]
a = weighted_test(weights)
x= list(range(len(a)))
plt.plot(x,a)
plt.grid()
plt.show()
```

```python
faces_of_die = [1, 2, 3, 4, 5, 6]
weights = [1, 1, 1, 1, 1, 1]

outcomes = []
n = 6
for _ in range(n):
    outcomes.append(weighted_choice(faces_of_die, weights))

c = Counter(outcomes)
for key in c:
    c[key] = c[key] / n
    
print(sorted(c.values()), sum(c.values()))
```
